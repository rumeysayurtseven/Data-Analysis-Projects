# Aggregate Statistics 

```python
# maximum number of Total Confirmed cases in each state 
grp = data.groupby("Name of State / UT")
```

```python
print(grp["Total Confirmed cases"].max())
```

```python
# sum of total number of confirmed cases in each state
grp=data.groupby('Name of State / UT')
```

```python
print(grp["Total Confirmed cases"].agg(sum))
```

```python
#in a short way
data.groupby('Name of State / UT')['Total Confirmed cases'].sum()
```

```python
#Sort the Total number of confirmed cases in each state/ UT in Descending orde
#In which State or UT maximum number of confirmed cases were recorded?
data.groupby("Name of State / UT")["Total Confirmed cases"].sum().sort_values(ascending = False)
```

```python
#group on multiple columns
data.groupby(['Date'])['Death','Total Confirmed cases','Cured/Discharged/Migrated','Name of State / UT']
```

```python
data.groupby(['Date'])['Death','Total Confirmed cases','Cured/Discharged/Migrated','Name of State / UT'].max()
```

```python
#filter the data first
newdata = data[data["Name of State / UT"] == "Maharashtra"]
```

```python
newdata.groupby("Total Confirmed cases")["Date"].max().sort_values(ascending = False).head(5)
```
