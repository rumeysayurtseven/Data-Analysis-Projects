6-Her kıta için şarap tüketiminin betimsel istatiğini yapınız.

```python
df.groupby("continent").describe()
```

7-Her kıta için ve her alkol cinsi için ortalama alkol tüketimini gösteriniz. 

```python
df.groupby("continent").mean()
```

8-Her kıta için ve aşağıdaki alkol cinsleri için belirtilen hesaplamaları gerçekleştiriniz.

beer_servings - std
spirit_servings - median 
wine_servings - min, mean, max
total_litres_of_pure_alcohol - mean, countbeer_servings

```python
df.groupby("continent").aggregate({"beer_servings": ["std", "count"], "spirit_servings": ["median"], "wine_servings":["min","mean", "max"],
                                  "total_litres_of_pure_alcohol": ["mean"]})
```

9- Her alkol cinsinin en çok tüketildiği ülkeyi bulun. 

```python
print(df.loc[df["beer_servings"].idxmax()][0])
print(df.loc[df["spirit_servings"].idxmax()][0])
print(df.loc[df["wine_servings"].idxmax()][0])
```
